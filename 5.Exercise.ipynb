```python
# first lets create a function to define the release year range
def categorize_release_year(release_year):
    if release_year < 2000:
        return "Vanha"
    elif release_year < 2020:
        return "Ei niin vanha"
    else:
        return "Uusi"
    
# We can use the .apply() method to create a new column 'release_year_group' based on the 'release_year' column.
df['release_year_group'] = df['release_year'].apply(categorize_release_year)
df.head()
```
