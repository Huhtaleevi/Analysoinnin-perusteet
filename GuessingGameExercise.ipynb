```python
nimi = "Leevi"
arvaus = ""
arvaustenMaara = 0

while True:
    arvaus = input("Please, guess my name.")
    arvaustenMaara += 1

    if arvaus == nimi:
        print("Congratulations!")
        print("Guesses:", arvaustenMaara)
        break

    else:
        vaaraArvaus = input("Do you want to quit (y/n)?")
        if vaaraArvaus == "y":
            print("Guesses:", arvaustenMaara)
            break
```
