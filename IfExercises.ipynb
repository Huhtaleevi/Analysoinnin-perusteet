```python
#Tehtävä 1
number1 = 20
number2 = 10

#a
if number1 == number2:
    print("Numbers are equal")

#b
if number1 > number2:
    print("Number1 is greater than number2")

#c
if number1 >= number2:
    print("Number1 is greater or equal to number2")

#d
if number1 != number2:
    print("Numbers are not equal")


#Tehtävä 2
number1 = 5
number2 = 3
number3 = 4

#a
if number1 and number2 == number3:
    print("Numbers are equal")

#b
elif number1 == number2 or number2 == number3:
    print("Numbers are equal")

#c
elif number1 > number2 and number1 > number3:
    print("Number1 is greater")

#d
if number1 > number2:
    print("Number1 is greater")

elif number2 > number3:
    print("Number2 is greater")

#e
if number1 == number2:
    print("Number1 and number2 are equal")

elif number1 == number3:
    print("Number1 and number3 are equal")


#Tehtävä 3
name1 = "Leevi"
name2 = "Tuomas"
name3 = "Iiro"

#a
if name1 == name2:
    print("Names are equal.")

#b
if name1 != name2:
    print("Names are not equal")

#c
if name1 == name2:
    print("Name1 and name2 are equal")

elif name1 == name3:
    print("Name1 and name3 are equal")












```
